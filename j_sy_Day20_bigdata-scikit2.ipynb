```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
bos = load_iris() # 일단 급한대로 아이리시 데이터를 boston으로 이름지음
```

```python
bos.data
```

```python
boston = pd.DataFrame(bos.data)
```

```python
boston
```

```python
bos.feature_names
```

```python
bos.target
```

```python
boston = pd.DataFrame(bos.data, columns = bos.feature_names)
```

```python
boston
```

```python
boston_target = pd.DataFrame(bos.target, columns = ['target'])
```

```python
boston_target
```

```python
boston = pd.concat([boston, boston_target], axis = 1)
```

```python
boston
```

```python
# 데이터를 training set, test set으로 나누어 모델을 설정하고 검증을 한다.

# 현재 수집된 데이터는 오류가 없다는 가정하에 위의 과정을 진행함
```

```python
from sklearn.model_selection import train_test_split
# training data set을 나눠주는 기능
```

```python
X_train, X_test, y_train, y_test = train_test_split(boston.iloc[:,:-1], boston.iloc[:,-1], random_state = 12)

# train_test_split으로 데이터셋을 4가지로 unpacking 해야 한다.
# train_test_split에서 test_size의 default 값이 0.33이므로 0.67이 training set, 0.33이 test_size이다.

# iris data는 개수가 150개밖에 안되서 학습시키는 training set에는 없는 데이터가 test set에 있을 가능성이 있다.  
# 이를 data leakage 문제라고 함 -> 즉 제대로 예측하지 못한다.

# training set, test set을 정하는 무작위 방식은 random_state 파라미터에 전달되는 인자(=seed)에 따라 다르다.
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train, y_train)
```

```python
knn.predict(X_test)
```

```python
y_test.values # pandas의 값을 numpy로 바꿔주려면 .values 하면 된다.
```

```python
knn.predict(X_test) == y_test.values # 원소마다 각각 비교해주네
```

```python
import numpy as np
```

```python
np.mean(knn.predict(X_test) == y_test.values) 
# True = 1, false = 0
# 97% 예측 성공률

# 이는 우아한 방식이 아니다.
```

```python
knn.score(X_test, y_test) 
# 예측률을 score로 볼 수 있다.
# 위와 같은 결과를 보임
```

```python
# 선형방식인 logistic regression 사용 
```

```python
lr = LogisticRegression()
```

```python
# cross_validation -> data를 n등분해서 training set, test set을 번갈아가면서 수행하면서 n개의 모델을 생성한다.

# 속도가 느리지만, 모든 data를 한번 이상씩 사용하므로 data leakage 문제가 발생하지 않는다.
# 데이터가 '작을 때' 사용하기에 좋다. (여러 번 반복해서 데이터의 양을 늘려주는 효과)

# 그러나 cross_validation은 최종모델이 아니다. 단지 모델의 성능 체크를 확인하기 위한 용도로 쓰인다. (평균치를 최종 모델로 사용할 수 없다.)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(KNeighborsClassifier(), boston.iloc[:,:-1], boston.iloc[:,-1], cv = 10, n_jobs=-1)
# estimator : 알고리즘을 instance화 한 것을 의미
# cross_validation은 데이터를 쪼개서 넣는게 아니라 전체 X, y를 전달하면 된다. - 전체 데이터를 n등분해서 알아서 돌아가기 때문
# cv -> 몇등분해서 할 것인가
# n_jobs = -1-> 최대의 코어를 활용
```

```python
%time np.mean(cross_val_score(KNeighborsClassifier(), boston.iloc[:,:-1], boston.iloc[:,-1], cv = 10, n_jobs=-1))
```

```python
# 그냥 부가적으로 설명하신 내용
from sklearn.svm import SVC # (SV classfication)

# 수학자가 만들어서 전부 수학으로 이루어져있음
# 속도는 느리지만 성능은 나름 좋게 나옴 -> 정확도는 높지만, 실무에 쓰기는 힘들다. (속도가 너무 느림)
```

```python
from sklearn.model_selection import GridSearchCV # CV -> cross validation 방식을 사용한다는 의미
```

```python
param_grid = {'n_neighbors': [1,2,3,4,5,6,7,8,9]}
```

```python
grid = GridSearchCV(KNeighborsClassifier(), param_grid)
# param_grid : hyper parameter를 gridsearch 내부에서 어떤 값을 사용할 건지 dictionary 형태로 전달하여 정해줄 수 있다.
```

```python
grid.fit(boston.iloc[:,:-1], boston.iloc[:,-1])
```

```python
grid.best_params_ # n이 5였을때 가장 성능이 좋다는 것을 알 수 있다.
```

```python
grid.best_score_ # 성능이 가장 좋았을 때의 score
```

```python
# --> 즉 일일이 hyper parameter를 바꿔가며 성능을 직접 비교할 필요가 없다. -> 가장 최적의 hyper parameter를 찾아내기 쉽다.
# 이 좋은 기능을 scikit에서 제공하기 때문에 scikit을 사용하는 가장 큰 이유가 된다.
```

```python
pd.DataFrame(grid.cv_results_).T

# pandas로 변형하여 hyper parameter 값에 따라 바뀐 여러 값들을 모두 보여 준다. (측정속도, 점수 등등)
# GridSearchCV -> 알고리즘 짜서 모듈로 만들자. 모든 알고리즘에 응용가능해서 효과적이다.
```

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris # target이 문자열로 되어있음 -> 숫자로 encoding 해야 함
```

```python
# pandas는 apply, map으로 못하는게 없다.
```

```python
iris.species.map({'setosa' : 0, 'virginica' : 1, 'versicolor' : 2})  

# 문자열을 0,1,2 처럼 숫자값으로 바꿔주는 이러한 방식을 label encoding 이라 한다.
# 숫자값을 바꿔준다는 것은 즉 각각이 숫자 크기 비교가 가능해지므로 X 에는 이를 적용하면 안된다. y(target)에만 적용하자
# label encoding은 categorical인 target에 사용하기에 적합하다.

# map에서는 dictionary 또는 함수를 사용할 수 있다.
```

```python
from sklearn.preprocessing import LabelEncoder # class -> 인스턴스화

# scikit에서 label encoding을 자동으로 해주는 기능
```

```python
le = LabelEncoder()
```

```python
le.fit_transform(iris.species) # label encoding을 자동으로 해준다.
```

```python

# 두번째 encoding 방식 - one hot encoding
# dummy 변수를 만든다고도 한다
```

```python
import pandas as pd
```

```python
pd.get_dummies(iris.species) # pandas에서 dummy 만드는 방식 - one hot encoding
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
ohe.fit_transform(iris[['species']]) # one hot encoder는 2차원을 전달해야됨 -> fency indexing 사용

# scikit에서 dummy 변수 만드는 방법 - one hot encoding
```

```python
ohe.fit_transform(iris[['species']]).toarray()
```

```python
ohe.inverse_transform([[1.,0.,0.]])
```

```python
le.inverse_transform([2])
```

```python
# 위의 두 코드가 pandas말고 scikit을 사용하는 이유이다.
# pandas는 encoding한 것이 원래 무엇이였는지 안알려주는데, scikit은 알려준다.
```

```python
# one hot encoding의 단점 : 차원을 늘려주기 때문에 데이터가 충분하지 않으면 성능이 안 좋게 나온다.

# label encoding의 단점인 숫자가 일정한 구조나 순서를 가지는 것
# we might confuse our model into thinking that a column has data with some kind of order or hierarchy, when we clearly don’t have it.

# 이러한 문제 때문에 혼동될 수 있다. 특히 X(결과 아닌 데이터들)는 label encoding보다 one hot encoding이 좋다. 
```
