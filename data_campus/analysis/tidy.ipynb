```python
weather = pd.read_csv('./tidy-data-master/data/weather.txt', sep='\t',na_values='-')
```

```python
weather.head()
```

```python
weather = weather.melt(weather.columns[:4], var_name='date').dropna()
```

```python
weather.describe()
```

```python
weather.drop(columns=['year','id'], inplace=True)
weather.head()
```

```python
weather.reset_index(drop=True, inplace=True)
```

```python
weather.date = weather.date.str.replace('d','')
```

```python
weather.head()
```

```python
weather.date = weather.date.astype('int64')
```

```python
weather.sort_values('month', inplace=True)
weather.reset_index(drop=True, inplace=True)
```

```python
weather.head()
```

```python
weather.info()
```
