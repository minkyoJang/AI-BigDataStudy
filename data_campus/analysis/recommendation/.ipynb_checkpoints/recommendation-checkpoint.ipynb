```python
reco_corr = recommend.T.corr()
```

```python
reco_corr.loc[1].sort_values(ascending=False).iloc[1:5] # corr이 높은 4명 선택하자
```

```python
data.info()
```

```python
def recoomendation_user(user_id, k, n=10):
    s = reco_corr.loc[user_id].sort_values(ascending=False).iloc[1:k+1].index # user_id만 뽑힘
    temp = pd.DataFrame(columns=['user_id', 'item_id','rating'],dtype='int64')
    for j, i in enumerate(s, start=1):
        t = set(data[data['user_id']==i].item_id) - set(data[data['user_id']==user_id].item_id) # 나랑 가장 비슷한 놈 - 내가 본 영화
        k = data[(data.item_id.isin(t)) & (data.rating==5) & (data.user_id==i)]
        k.rating = k.rating * (1-0.05*j)
        temp = pd.concat([temp, k])
    else:
        return temp.rename(columns={'item_id':'movie_id'}).merge(item[['movie_id','title']]).drop_duplicates().groupby('title').mean().sort_values('rating',ascending=False)[:n]
#     .index
```

```python
recomm = recoomendation_user(42,5)
```
