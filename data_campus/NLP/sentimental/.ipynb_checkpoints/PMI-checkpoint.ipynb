```python
import sqlite3
```

```python
con = sqlite3.connect(
    'E:/data_campus/data_processing/unstructuredData/news.db')
cur = con.cursor()

cur.execute('''
    SELECT content
    FROM table2
    ORDER BY id ASC
''')
corpus = [_[0] for _ in cur.fetchall()]
```

```python
from konlpy.tag import Kkma, Hannanum
from collections import defaultdict
```

```python
pos_tag = Kkma().pos

TDM = defaultdict(lambda: defaultdict(int))

for i, d in enumerate(corpus):
    for t in [_[0] for _ in pos_tag(d)
              if _[1].startswith('N') or _[1].startswith('V')
              if len(_[0]) > 1]:
        TDM[t][i] += 1
```

```python
pos_tag = Hannanum().pos

hanTDM = defaultdict(lambda: defaultdict(int))

for i, d in enumerate(corpus):
    for t in [_[0] for _ in pos_tag(d)
              if _[1].startswith('N') or _[1].startswith('V')
              if len(_[0]) > 1]:
        hanTDM[t][i] += 1
```

```python
sorted([{t: len(docList)} for t, docList in TDM.items()],
       key=lambda x: sum(x.values()), reverse=True)
# positive : '노력', '한국', '평화'
# negative : '규제', '일본', '비판'
```

```python
sorted([{t: len(docList)} for t, docList in hanTDM.items()],
       key=lambda x: sum(x.values()), reverse=True)
# positive : '노력', '한국', '강화'
# negative : '수출규제', '일본', '비판'
```

```python
from math import log2
pSeed = ['노력', '한국', '평화']
nSeed = ['규제', '일본', '비판']

N = len(corpus)

pSeeddf = list(TDM[pSeed[0]])  # `노력`이 등장하는 문서의 인덱스 리스트
nSeeddf = list(TDM[nSeed[0]])

pProbY = len(pSeeddf) / N  # `노력`은 전체 N개 문서 중 몇 개 문서에서 등장?
nProbY = len(nSeeddf) / N

posPMI = defaultdict(float)
negPMI = defaultdict(float)

posNPMI = defaultdict(float)
negNPMI = defaultdict(float)

for t in TDM:
    termdf = list(TDM[t])
    pX = len(termdf)/N
    posXY = (len(set(pSeeddf).intersection(termdf))+1) / N
    negXY = (len(set(nSeeddf).intersection(termdf))+1) / N

    posPMI[t] = log2(posXY / (pX*pProbY))
    negPMI[t] = log2(negXY / (pX*nProbY))

    posNPMI[t] = posPMI[t] / -log2(posXY)  # 정규화, -1~1로 만든다
    negNPMI[t] = negPMI[t] / -log2(negXY)
```

```python
SO = defaultdict(float)
nSO = defaultdict(float)
for t in TDM:
    SO[t] = posPMI[t] - negPMI[t]
    nSO[t] = posNPMI[t] - negNPMI[t]
```

```python
pSeed = ['노력', '한국', '평화']
nSeed = ['규제', '일본', '비판']

N = len(corpus)

posPMI = defaultdict(lambda: defaultdict(float))
negPMI = defaultdict(lambda: defaultdict(float))

posNPMI = defaultdict(lambda: defaultdict(float))
negNPMI = defaultdict(lambda: defaultdict(float))
for i in range(len(pSeed)):
    pSeeddf = list(TDM[pSeed[0]])  # `노력`이 등장하는 문서의 인덱스 리스트
    nSeeddf = list(TDM[nSeed[0]])

    pProbY = len(pSeeddf) / N  # `노력`은 전체 N개 문서 중 몇 개 문서에서 등장?
    nProbY = len(nSeeddf) / N

    for t in TDM:
        termdf = list(TDM[t])
        pX = len(termdf)/N
        posXY = (len(set(pSeeddf).intersection(termdf))+1) / N
        negXY = (len(set(nSeeddf).intersection(termdf))+1) / N

        posPMI[pSeed[i]][t] = log2(posXY / (pX*pProbY))
        negPMI[nSeed[i]][t] = log2(negXY / (pX*nProbY))

        posNPMI[pSeed[i]][t] = posPMI[pSeed[i]][t] / -log2(posXY)  # 정규화, -1~1로 만든다
        negNPMI[nSeed[i]][t] = negPMI[pSeed[i]][t] / -log2(negXY)
```

```python
SO = defaultdict(float)
nSO = defaultdict(float)

for t in TDM:
    for i in range(len(pSeed)):
        SO[t] += posPMI[pSeed[i]][t] - negPMI[pSeed[i]][t]
        nSO[t] += posNPMI[pSeed[i]][t] - negNPMI[pSeed[i]][t]
```

```python
sorted(nSO.items(), key=lambda x:x[1], reverse=True)[:10],\
sorted(nSO.items(), key=lambda x:x[1])[:10]
```

```python
pos_tag = Kkma().pos

for i, d in enumerate(corpus):
    docSO = 0.0
    docNSO = 0.0
    for t in [_[0] for _ in pos_tag(d)
              if _[1].startswith('N') or _[1].startswith('V')
              if len(_[0]) > 1]:
        docSO += SO[t]
        docNSO += nSO[t]
        
    print('{0}번째 문서 : SO={1}|{2}, NSO={3}|{4}'.format(
    (i+1), docSO, 1 if docSO > 0 else 2,
    docNSO, 1 if docNSO >0 else 2)
         )
        
```
