# 판다스의 역할
- 판다스는 넌파이 기반으로 eda (exp..)하고 데이터 전처리 하는 애
- 즉 EDA 하는거에요. 탐색적 데이터 분석. 
- 기술 통계 편하게 하는 것
- 기초 분석하는 도구 
- 데이터 먼징 렁글링 멍징 하는 도구
  
#### eda?
    - Exploratory Data Analysis
    - 미싱 데이터를 찾는다.
    
#### 넌파이 역할:벡터기반으로 데이터를 처리하는 라이브러리!
    - 이 동네는 따라서 벡터 기반인데 우리는 어레이 기반이라고 할게요.
