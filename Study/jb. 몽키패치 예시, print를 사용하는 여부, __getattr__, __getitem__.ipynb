.으로 접근했을 때 그 뒤에것이 없으면   \_\_getattr__()을 실행한다

```python
class X:
    def __getattr__(self,attr):
        return '문근영'
```

```python
a = X()
```

```python
dir(a)
```

```python
a.a
```

```python
a.v
```

```python
a.x
```

```python
a.f
```

```python
class X:
    def __getattr__(self,attr):
        if attr == 'a':
            return '문근영2'
        return '문근영'
```

```python
a = X()
```

```python
a.a
```

```python
a = 1.0
```

```python
a.hex()
```

- 값이 없을 때 \_\_getattr\_\_ 사용 (attribute error가 발생할 때)
- 값이 있을 때 getattr() 사용해서 실행시켜준다.

```python
getattr(a, 'hex')()
```

```python
__getattribute__ # .으로 탐색했을 때 사용할 수 있는것
```

try는 중첩된 부분에 에러가 나면 또 에러가 난다 getattr 또한 그렇다.

```python
try:
    a = 1/0
except:
    a = 2/0
```
