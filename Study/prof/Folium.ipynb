```python
import pandas as pd
import random

#Iterable
list_data = [10, 20, 30, 20, 15, 30, 45]

#Dicts of iterables
cat_1 = ['y1', 'y2', 'y3', 'y4']
index_1 = range(0, 21, 1)
multi_iter1 = {'index': index_1}
for cat in cat_1:
    multi_iter1[cat] = [random.randint(10, 100) for x in index_1]

cat_2 = ['y' + str(x) for x in range(0, 10, 1)]
index_2 = range(1, 21, 1)
multi_iter2 = {'index': index_2}
for cat in cat_2:
    multi_iter2[cat] = [random.randint(10, 100) for x in index_2]
```

```python
!pip install vincent
```

```python
import vincent
bar = vincent.Bar(multi_iter1['y1'])
bar.axis_titles(x='Index', y='Value')
bar.to_json('asset/vis1.json')
```

```python
line = vincent.Line(multi_iter1, iter_idx='index')
line.axis_titles(x='Index', y='Value')
line.legend(title='Categories')
line.to_json('asset/vis2.json')
```

```python
scatter = vincent.Scatter(multi_iter2, iter_idx='index')
scatter.axis_titles(x='Index', y='Data Value')
scatter.legend(title='Categories')
scatter.colors(brew='Set3')
scatter.to_json('asset/vis3.json')
```

```python
area = vincent.Area(list_data)
area.to_json('asset/vis4.json')
```

```python
seoul_topo_json = open('asset/seoul_municipalities_topo_simple.json', encoding='utf8')
seoul_topo_json = json.load(seoul_topo_json)

seoul_geo_json = open('asset/seoul_municipalities_geo_simple.json',encoding='utf-8')
seoul_geo_json = json.load(seoul_geo_json)
```

```python
m9 = folium.Map( location=[37.5665, 126.9780], tiles='Mapbox Bright') 

folium.GeoJson(seoul_geo_json,name='geojson').add_to(m9)

folium.TopoJson(seoul_topo_json,"objects.seoul_municipalities_geo",name='topojson').add_to(m9)

folium.LayerControl().add_to(m9)
m9
```

```python
data = pd.DataFrame.from_dict(seoul_geo_json['features']).properties
# data = pd.DataFrame.from_dict(seoul_topo_json['objects']['seoul_municipalities_geo']['geometries']).properties
```

```python
keys = data[0].keys()
```

```python
for i in keys:
    print(i)
```

```python
data_list = {}
for key in keys:
    temp_list = []
    for inst in data:
        temp_list.append(inst[key])
    data_list[key] = temp_list    
```
