Pedro Domingos : 유명한 사람

- bias: 결과와 떨어진 정도
- variance(분산, 흩어진 정도): 결과들이 퍼져있는 정도 
- bias가 작을수록 variance 작을 수록 좋다.
    - 둘 다 작기는 쉽지 않다. trade-off 관계에 있으므로 하나가 커지면 하나가 작아지는 관계이다.
    - 적절한 모델을 찾는 것이 중요하다.

#### 모델이 복잡해지면
- bias가 작아진다.
- variance가 커진다.
- 둘을 합쳐서 가장 작아지는 모델을 찾아야한다. -> model selection

학습시킬수록 bias가 줄고 Training error가 작아지지만 variance가 높아져서 Test Error가 커진다.

get_params : 내가 바꿀 수 있는 하이퍼 파라미터들 -> 바꿔서 성능을 바꿀 수 있다.
- 주의 : 성능의 종류가 여러가지이다.
- 정확도 높일 때도 사용되고 속도도 포함된다. cs 관점으로 본 것.
- n_jobs(number of ~): 몇개의 코어를 사용해서 학습시킬 건지 -> 동시에 여러개 코어를 했으니 속도가 빠르다 -> 예측할 때 빨리 하는 것도 성능이다.

```python
knn2.get_params()
```

```python
knn2 = KNeighborsClassifier(n_neighbors=3, n_jobs=-1) # n_jobs=-1은 자동 : 최고로 많이 사용해서 계산 
```
