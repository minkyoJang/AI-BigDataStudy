```python
def recommendation_user(user_id, k,n=10): #user_id에 대해서 근처 5개 knn해주고, n개 추천해주는데 디폴트는 10개
    s = reco_corr.loc[user_id].sort_values(ascending=False)[1:k+1].index #나하고 비슷한애 찾아요. 근처 몇개. 인덱스 값이 나와
    temp = pd.DataFrame(columns=['user_id','item_id','rating']) #값을 템프를 미리 만들어 줍니다. 그리고 여기에 추가할거야
    for j, i in enumerate(s,1): #i는 유저에 대한거 
        t = set(data[data.user_id==i].item_id)  - set(data[data.user_id==user_id].item_id)# 내가 본거랑 그녀석 본거 빼줌
        k = data[(data.item_id.isin(t)) & (data.rating==5) & (data.user_id==i)]
        k.rating = k.rating * (1-0.05*j)#밑에있을수록 005씩 가중치 줄여서 붙이기. j 0qnxj 
        temp = pd.concat([temp, k]) #concat은 테이블 붙이는거. 
    else:
        return temp.rename(columns={'item_id':'movie_id'}).merge(item[['movie_id','movie_title']]).drop_duplicates().groupby('movie_title').mean().sort_values('rating',ascending=False)[:n].index
    # 머지는 이너조인 같아. 칼람 합쳐줘. 
    # 듀플리케이트는 중복여부 찾게해주고, drop_duplicates로 중복 제거
    # 그룹바이로 묶는데 무비 이름으로 묶고, 평점의 평균으로 뺴내
```
