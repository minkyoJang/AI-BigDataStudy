## (1) 키 스트럭쳐
- 메모리 상태에다가 리스트로 저장하는 형식
- 백오브워드로하면서 공간 줄이고 빠르게 탐색가능케
- 일반적으로 언스트러겨드로부터 정보 추출 탐색할떄 사용가능한 구조

```python

D= [["A","B"],["C","D"], ["A","C"] ]


Dictionary= defaultdict(lambda: -1) #-1값을주는건 다음 위치가 어디가 되는지 담는 구조라서
# 다음 위치가 -1이라는건 없다는것. 없는 상태에서 주소는 항상 0부터 최소 시작하는 양수니까 더이상 없단뜻
Postings= list() #일단 메모리에서 작업하고 파일의 형태로 쪼갤거에요

V=list() # 단어목록을 뽑아야 하니 v를 뽑아야해

#다큐먼트 집합 콜렉션 열면 텀이 나올 거고
for j,d in enumerate(D): 
    for t in d:
        if t not in V: #만일 포함안된건 새로운 키값이 나온거니까 어팬드
            V.append(t) #어팬드했으니 몇 번째 ㅣㄴ덱스인지 나올거
        i = V.index(t)
        #포스팅 작업을 해줄거에요
        
        #몇번쨰 문서에서 몇 번 나왔는지를 포스팅에 저장할건데. 일단 1이라고 가정하고 시작해보죠
        #안에 넣을 것은 튜플로 만듭시다.
        #_struct= (j,1,?) #(j번쨰 문서에서 몇 번 나왔고, 다음 위치는 어디다)
        ## 다음 위치를 어떻게 하냐면
        nextPosition= Dictionary[i] #뭐라도 값이 있음 값이 있고, 없음 -1 나왔을 거니까
        _struct= (j, 1, nextPosition)
         #딕셔너리의 아이를 바꿀거에요. 포스팅의 현재 위치로
            ## 현재 위치하면 랭쓰하면 나와요. 처음엔 아무것도 없으니 0이고 , 거기엔 _struct있지
            ## _struct 따라가면 다음이 어딘지 알 수 있지
        Dictionary[i]=len(Postings)
        Postings.append(_struct)
        
```

```python
from collections import defaultdict
D= [["A","B"],["C","D"], ["A","C"] ]

Dictionary= defaultdict(lambda: -1) 
Postings= open("postings.dat","w")
V=list() 

for j,d in enumerate(D): 
    for t in d:
        if t not in V: 
            V.append(t) 
        i = V.index(t)
        nextPosition= Dictionary[i] 
        _struct= (str(j), str(1), str(nextPosition))
        Dictionary[i] = Postings.tell() #위치 반환하라고 할거구요
        Postings.write(",".join(_struct)+"\n") #이거 저장하라고 할거에요
                  
Postings.close()
```

## (3)struct 패키지 이용해서 진행 
- pack&unpack 바이트어레이로 변환하여 진행
- 마치 C에서 쓰는 것처럼 바이트로 저장했다가 불러와서 그대로 사용할 수 있쬬
- 다시 위에것 복사해서

```python
from collections import defaultdict
from struct import pack, unpack
D= [["A","B"],["C","D"], ["A","C"] ]

Dictionary= defaultdict(lambda: -1) 
Postings= open("postings.dat","wb") #바이트로 저장하니까 여기 바뀌고
V=list() 

for j,d in enumerate(D): 
    for t in d:
        if t not in V: 
            V.append(t) 
        i = V.index(t)
        nextPosition= Dictionary[i] 
        #pack(포맷이 나오고, 값이 나와요)
        _struct= pack("iii", j, 1, nextPosition) #이제 스트링 필요없어서 이렇게 쓰고, pack으로 변환할거에요
        #"iii" 나는 인티저 3개를 입력하겠다
        Dictionary[i] = Postings.tell() #위치 반환하라고 할거구요@@
        Postings.write(_struct) # 변환
                  
Postings.close()
```

```python
from string import punctuation

from konlpy.tag import Kkma
import re
from string import punctuation
from nltk.tokenize import word_tokenize

#word_tokenize(corpus[0])
## 구두점 날리기
pattern1=re.compile(r"[{0}]".format(re.escape(punctuation)))
pattern1.sub("",corpus[0]) #따옴표 점 같은애들 다 날릴거에요


import re
from string import punctuation
from nltk.tokenize import word_tokenize

#word_tokenize(corpus[0])
## 구두점 날리기
pattern1=re.compile(r"[{0}]".format(re.escape(punctuation)))
pattern1.sub("",corpus[0]) #따옴표 점 같은애들 다 날릴거에요


## 영어 5글자이상이면 날려요_이메일 날라감

pattern2=re.compile(r"[A-Za-z]{5,}")

pattern2.sub("",pattern1.sub("",corpus[0]))


## 패턴3: 특수문자 날리기 띄어쓰기 빼고
pattern3=re.compile(r"[^ 0-9A-Za-z가-힣]")
pattern3.sub("", pattern2.sub("",pattern1.sub("",corpus[0])))

#  \s는 스페이스 즉 2개 이상 띄어져있으면 처리
pattern4=re.compile(r"\s{2,}")
pattern4.sub(" ", pattern3.sub(" ", pattern2.sub(" ",pattern1.sub("",corpus[0]))))

collection= list()

for _ in corpus:
    collection.append(pattern4.sub(" ", pattern3.sub(" ", pattern2.sub(" ",pattern1.sub("",_)))).strip()) 
    #스트립으로 시작 양 옆 공백있으면 잘라줌
```

```python
V=list() #전체 보캐뷸러리 셋
D=list()

DTM= defaultdict(lambda:defaultdict(int))
for d in collection: #collection은 디비에 저장해둔. 그 정제한 뉴스기사
    # 전처리(토크나이즈, 레마타이크_어근추출_여기서 한거, 컨트롤드 보케불러리_명사만 갖고 등등 했죠)
    i=len(D)
    D.append(i)
    #명사 추출
    kkma=Kkma()
    for t in [_[0] for _ in kkma.pos(d) if _[1].startswith("N")]: #수많은 명사 중에서 N으로 시작하는애
        # 포스 돌리면 여기에 "형태소"-"품사"나오는 구조인데, 우리는 품사를 가지고 거를거니까 [1]번째
        # 집어넣는건 [0]번쨰 명사만
        # 두가지 방식 있음(브이 먼저 만들거나. 근데 순서대로 해보죠.)
        if t not in V: #추가를 해야하닊
            V.append(t) #어팬드 하니 제이를 구할수도 있고 그건 얘의 인덱스. 아이와 제이 구했으니 dtm 만들죠
        j=V.index(t)
        DTM[i][j] +=1
            # 아이번쨰 문서에 제이 문자가 몇 번나왔는지 디티엠에 저장
```

### TF normalization
- 로그를 씌워서 숫자가 많이 크지 않게 해요. 10나온거나 100나온거나 차이가 로그 씌우면1,2정도니까 그럼 숫자가 많이 커도 무리없겠죠
- 최소 1이상인 값들이 나오게
    - tf(td)= 1+log⁡𝑓(𝑡,𝑑), 𝑖𝑓 𝑓(𝑡,𝑑)>0
    - 𝑡𝑓(𝑡,𝑑)=𝛼+(1−𝛼)(𝑓(𝑡,𝑑))/max┬𝑡⁡〖𝑓(𝑡,𝑑)〗 
        - 그러면 항상 0~1사이의 값을 갖는데 0부터 말고 특정 알파부터 시작할 수 있게끔 할 수 있어요
        - 특정경우랑 아닌경우.
            - 바이너리 클래시파이를 만들떄 이런식의 것들 많이 만들죠. 특히 로지스틱에서도 1- 어쩌구.
            - 그래서 오즈라고 하는 승산계산한것처럼. 우리도 이렇게 해요
            - 저게 일반적으로 많이 쓰는 티에프에요.
            
<> tf는 0~1사이 스케일링, idf는 가중치이며 보통 이게 중요하게 쓰여요.            
            
### IDF weighting
- 𝐼𝐷𝐹(𝑡)=1+log⁡(𝑁/(𝑑𝑓(𝑡)))
- 인버티드 다큐먼트. 
- 만약에 디에프를 섞어서 쓰면 문서에서 많이 나온거 가중치를 받는거니까 왼쪽에 해당하는거고, 적게나올수록 가중치가 많이 나와야 하니까 인버티드 다큐먼트 프리퀀시
- 여기보면 로그 디에프 특정 텀이 들어간 프리퀀시에요. 
- 그 ㅈㅇ에서 랑지엔이니까 콜랙션. 랭쓰 사이즈 즉 다큐먼트 사이즈. 즉 전체문서에서 다 나왔다는게 엔이 되고 분자도 엔이되고 로그를 씌우니까 0이 되는데 이거 스무딩 해주려고 1을 더해주는거죠.

### idf 종류

```python
N=len(D)
idf1= lambda nt:log(N/nt)
idf2= lambda nt:log(1+N/nt)
idf3= lambda mnt, nt: log(mnt/(1+nt)) #mnt: max nt
idf4= lambda nt:log((N-nt)/nt)
idf5= lambda nt:log(1+(N-nt)/nt) #이건 항상 1보다 큰 양수만 나와요
```

### TF 종류

```python
for j, docs in TDM.items(): #j가 있으면 다큐먼튼 있을거고
    IDF1= idf1(len(docs))
    IDF2= idf2(len(docs))
    IDF4= idf4(len(docs)) #@@@@@@@@@@@@@@@@@@@@@@
    for i, f in docs.items():
        TF1=tf1(f,sum(DTM[i].values())) #TDM 구조에서는 조금 힘들어요. 왜냐하면 각각 문서 뒤져야하니까
            # 그래서 DTM을 만들었어요. DTM의 아이번쨰 문서에 해당하는게 프리퀀시고 그거 썸해서 보내주면 될거에요. 그게 티에프 일이에요
            # j번재의 TF, i번째의 문서에서
        TF2=tf2(f)
        TF3=tf3(f, max(DTM[i].values()),0)
        
        W1 = f*IDF1
        W2 = TF2 * 1
        W3 = TF2*IDF1
        W4= TF3*IDF1
        
        print(V[j], i, TF1, TF2, TF3) # 브이의 제이번째의 애가 가지는 그거는 하나(TF1) 둘(TF2) 셋(TF3)
        # 합리적 0 0.00476~~~   1.0~~  0.54~~
        # 합리적 30 0.0046~~~    0.6~~  0.54~~
    print(V[j], IDF1, IDF2, IDF4)
```
