### 앞서 원도 했으니, 다각형도 가능해요

```python
m7 = folium.Map( location=[37.5665, 126.9780], tiles='Stamen Terrain') 

folium.RegularPolygonMarker(
    [37.5665, 126.9780],
    popup='polygon1',
    fill_color='#132b5e',
    number_of_sides=3,
    radius=5
    ).add_to(m7)
folium.RegularPolygonMarker(
    [37.4665, 126.9780],
    popup='polygon2',
    fill_color='#45647d',
    number_of_sides=4,
    radius=10
    ).add_to(m7)
folium.RegularPolygonMarker(
    [37.5665, 126.8780],
    popup='polygon3',
    fill_color='#769d96',
    number_of_sides=6,
    radius=15
    ).add_to(m7)
folium.RegularPolygonMarker(
    [37.4665, 126.8780],
    popup='polygon4',
    fill_color='#769d96',
    number_of_sides=8,
    radius=20
    ).add_to(m7)
m7
```

    - 한 개는 키, 한 개는 벨류!
        - 키는 피쳐 프로텨 네임(이름 맞춰요). 저 키에 똑같은 값이 나올때 데이터 프레임에서 저 애를 나타나겠다.

```python
import pandas as pd
```

```python
m10 = folium.Map( location=[37.5665, 126.9780], tiles='Mapbox Bright') 

seoul_geo_df = pd.read_csv('asset/seoul_geo_df.csv')
m10.choropleth(
    geo_data=seoul_geo_json,
    name='choropleth',
    data=seoul_geo_df,
    columns=['name', 'code'],
    key_on='feature.properties.name',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Seoul_choropleth'
)
```
