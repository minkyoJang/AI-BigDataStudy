#### Regular EX 정규식

```python
import re
from nltk.tokenize import regexp_tokenize
pattern=re.compile(r"\b(\w+)\b") 
#양쪽에 워드바운더리 있으니 공백으로 구분되어있는 애들 사이의 문자(알파벳 대소문자, 숫자1개이상 )==어절
regexp_tokenize("You need Python!!!! 12 한#%@글", pattern)
```

@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@

```python
import matplotlib.pyplot as plt
from math import log


token= [_[0] for _ in sorted(text.vocab().items(), key=lambda x : x[1], reverse=True)]  #@@@
print(text.vocab().items())
print(sorted(text.vocab().items(), key=lambda x : x[1], reverse=True)) # 지금 튜플 타입으로 ('턱',1)있어
# 람다는 익명함수고, 현재 x는 파라미터, x[1]은 리턴값. 여기에서 x[1]은 '1'이라 이것들 정렬해서 소팅?!
freq= [_[1] for _ in sorted(text.vocab().items(), key=lambda x: x[1], reverse=True)] #@@@
_freq=[log(_) for _ in freq]
rank=[1/(_+1) for _ in range(len(token))] # 표기법@@
_rank=[log(_) for _ in rank]
plt.plot([i for i in range(len(token))], freq, "r-")
plt.plot([i for i in range(len(token))], _freq, "r-")
plt.plot([i for i in range(len(token))], rank, "b-")
plt.plot([i for i in range(len(token))], _rank, "b-")
plt.show()
```

```python
def findBigram(tokens):
    ngram= defaultdict(int) #@@@@@@@@@@@@@@@@@@@@@@@@@@뀨?왜 int가 들어가용?
    for token, freq in tokens.items(): #토큰과 프리퀀시 다 불러온대 이렇게 하면.@@@token과 freq는 걍 만든거?
        syllables= token.split()
        for i in range(len(syllables)-1):# 아까 빼기 1한다그랬으니까
            ngram[tuple(syllables[i:i+2])] += freq ##low 5개, lower는 2개니까 2개씩 묶었을떄 이 쌍의 총 토큰스에서 얼마있는지★★★
    return ngram
```

```python
def mergeMaxkey(ngram, tokens,K):
   # maxKey= max(ngram, key=lambda x:x[1])
    maxKey=max(ngram, key=ngram.get)
    maxValue=max(ngram.values())
    newTokens=dict()
    for token,freq in tokens.items():
        newTokens[re.sub(" ".join(maxKey),"".join(maxKey), token ) if maxValue > K else token] = freq
    return newTokens
```
